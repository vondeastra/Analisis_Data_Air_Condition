### Cleaning Data

#### Fixing ... problem

#### **PENANGANAN MISSING VALUE**

Saya akan menggunakan Interpolasi berbasis waktu dalam menangani missing value pada kategori polutan, hal ini dikarenakan data bersifat time series dan polusi berubah secara bertahap. Maka proses yang akan dilakukan:
- Membuat datetime
- Interpolasi Polutan
- Mengisi missing Value
- cek missing value kembali
- jika masih ada missing value, maka akan menggunakan teknik fillna 

```python
# Mmebuat Datetime

df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
df = df.sort_values(by=['station','datetime'])
df.head()
```

```python
# Interpolasi untuk mengisi missing value

cols_polutan = ['PM2.5','PM10','SO2','NO2','CO','O3']
df[cols_polutan] = df.groupby('station')[cols_polutan].transform(lambda x: x.interpolate(method='linear'))
```

```python
# Tangani missing pada cuaca
cols_cuaca = ['TEMP','PRES','DEWP','RAIN','WSPM']
df[cols_cuaca] = df.groupby('station')[cols_cuaca].transform(lambda x: x.fillna(method='ffill'))
```

```python
# Tangani missing wd
df['wd'] = df['wd'].fillna(method='ffill')
```

```python
#Cek kembali missing value setelah penanganan
df.isnull().sum()
```

```python
# Penanganan missing value yang tersisa dengan imputasi
df = df.fillna(method='bfill')
```

#### **Tambahan pembersihan**

```python
# Hapus tabel no

df = df.drop(columns=['No'])
```

**Insight:** (Opsional)
- Missing Value berhasil ditangani, dengan menggunakan metode interpolation pada polusi dan forward fill pada cuaca
- Outlier yang ditemukan valid dan tidak perlu dilakukan perubahan, dataset mampu menangkap kejadian ekstrem yang relevan untuk insight.
- distirbusi data antar lokasi seimbang dan tidak ada error data

### Pastikan Datetime

```python
df['datetime'] = pd.to_datetime(df[['year','month','day','hour']])
```

### Fitur waktu 

```python
# Mmebuat fitur tanggal
df['date'] = df['datetime'].dt.date
df[['datetime','date']].head()
```

```python
# Membuat nama bulan untuk visualisasi
df['month_name'] = df['datetime'].dt.month_name()
df[['month','month_name']].drop_duplicates().sort_values(by='month')
```

```python
#Tambah nama hari
df['day_name'] = df['datetime'].dt.day_name()
df[['datetime','day_name']].head()
```

```python
# Membuat year_month untuk analisis bulanan
df['year_month'] = df['datetime'].dt.to_period('M')
df[['datetime','year_month']].head()
```

```python
# Mmembuat Kolom time_period
def time_period(h):
    if 0 <= h < 6:
        return 'Dini Hari'
    elif 6 <= h < 12:
        return 'Pagi'
    elif 12 <= h < 18:
        return 'Siang'
    else:
        return 'Malam'

df['time_period'] = df['hour'].apply(time_period)

#Cek Kategori yang terbentuk
df['time_period'].unique()
```

```python
# Cek Mapping telah sesuai atau belum
df[['hour','time_period']].drop_duplicates().sort_values(by='hour')
```

### **Fitur Musim**

Untuk Menjawab Pertanyaan Bisnis yang telah kita desain

```python
# Mmebuat kolom musim
def season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Error'

df['season'] = df['month'].apply(season)

# Cek Kategori yang telah dibentuk
df['season'].unique()
```

```python
# Cek Mapping sudah benar atau belum
df[['month','season']].drop_duplicates().sort_values(by='month')
```

### **Membuat Kategori Kualitas Udara (PM2.5)**

```python
def pm25_category(x):
    if x <= 35:
        return 'Baik'
    elif x <= 75:
        return 'Sedang'
    elif x <= 115:
        return 'Buruk'
    elif x <= 150:
        return 'Tidak Sehat'
    elif x <= 250:
        return 'Parah'
    else:
        return 'Berbahaya'

df['pm25_category'] = df['PM2.5'].apply(pm25_category)

# Cek mapping sudah benar
df[['PM2.5','pm25_category']].head()
```

```python
# menentukan batas "Tidak sehat" untuk PM2.5
df['is_unhealthy'] = df['PM2.5'] > 90
df[['PM2.5','pm25_category','is_unhealthy']].head()
```

### Pertanyaan 2:

Seberapa besar perbedaan rata-rata PM2.5 antara musim dingin (Desember-Februari) dan musim panas (Juni-Agustus) pada setiap titik pemantauan selama periode 2013-2017

```python
season_station = (
    df[df['season'].isin(['Winter','Summer'])]
    .groupby(['station','season'])['PM2.5']
    .mean()
    .reset_index()
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=season_station,
    x='station',
    y='PM2.5',
    hue='season'
)

plt.title('Perbandingan Rata-rata PM2.5: Winter vs Summer per Station')
plt.xlabel('Titik Pemantauan')
plt.ylabel('Rata-rata PM2.5')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

**Insight:**
- Diseluruh Titik Pemantaun cenderung Winter > Summer
- Gap polusi yang dihasilkan per musim nampak sangat jauh

```python
pivot_season = season_station.pivot(
    index='station',
    columns='season',
    values='PM2.5'
)

pivot_season['difference'] = pivot_season['Winter'] - pivot_season['Summer']

pivot_season = pivot_season.sort_values(by='difference', ascending=False)
pivot_season
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=pivot_season.reset_index(),
    y='station',
    x='difference'
)

plt.title('Selisih PM2.5 (Winter - Summer) per Station')
plt.xlabel('Selisih PM2.5')
plt.ylabel('Titik Pemantauan')

plt.tight_layout()
plt.show()
```

**Insight:**
- Selisih gap yang terjadi antara musim winter dengan Summer di tiap titik pemantaun berkisar 23-43

## CLUSTERING ANALISIS (LANJUTAN)

### Manual Grouping

Manual grouping dilakukan untuk mengelompokkan titik pemantauan ke dalam tingkat risiko berdasarkan kombinasi indikator polusi, sehingga mempermudah interpretasi data dan membantu menentukan prioritas pengawasan kualitas udara. Pada tahap ini kita telah melakukannya, namun sekarang kita akan Mengelompokkan titik pemantauan (station) berdasarkan
- tingkat rata-rata polusi PM2.5
- Resiko Kesehatan

```python
# Kita terapkan Manual Grouping berdasarkan mean_PM25 dan unhealthy_ratio

station_summary = df.groupby('station').agg(
    mean_PM25=('PM2.5', 'mean'),
    median_PM25=('PM2.5', 'median'),
    unhealthy_ratio=('is_unhealthy', 'mean')
).reset_index()
```

```python
# Buat Aturan mengenai parameter PM2.5 dan unhealthy_ratio untuk menentukan risk group (resiko sakit lebih tinggi atau tidak)

def risk_group(row):
    if row['mean_PM25'] >= 85 and row['unhealthy_ratio'] >= 0.34:
        return 'High Risk'
    elif row['mean_PM25'] >= 75 or row['unhealthy_ratio'] >= 0.30:
        return 'Medium Risk'
    else:
        return 'Low Risk'

station_summary['risk_group'] = station_summary.apply(risk_group, axis=1)

station_summary = station_summary.sort_values(by='mean_PM25', ascending=False)
station_summary
```

```python
# Validasi Hasil Grouping dalam risiko polusi
station_summary['risk_group'].value_counts()
```

```python
# Visualisasi Klasifikasi Station Berdasarkan Risiko Polusi

plt.figure(figsize=(10,6))

sns.barplot(
    data=station_summary,
    x='station',
    y='mean_PM25',
    hue='risk_group'
)

plt.title('Klasifikasi Station Berdasarkan Risiko Polusi')
plt.xlabel('Station')
plt.ylabel('Rata-rata PM2.5')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Distribusi Kategori Risiko Station

plt.figure(figsize=(8,5))

sns.countplot(
    data=station_summary,
    x='risk_group',
    order=['High Risk','Medium Risk','Low Risk']
)

plt.title('Distribusi Kategori Risiko Station')
plt.xlabel('Kategori Risiko')
plt.ylabel('Jumlah Station')

plt.tight_layout()
plt.show()
```

**INSIGHT:**

Melalui manual grouping, titik pemantauan berhasil diklasifikasikan ke dalam tiga kategori risiko berdasarkan kombinasi tingkat polusi dan frekuensi kondisi tidak sehat. Kategori High Risk, Medium Risk, dan Low Risk merupakan hasil pengelompokan manual berdasarkan tingkat rata-rata PM2.5 dan frekuensi kondisi tidak sehat. Kategori ini digunakan untuk menyederhanakan interpretasi data dan membantu mengidentifikasi tingkat risiko kualitas udara pada setiap titik pemantauan. Kelompok High Risk menunjukkan area dengan paparan polusi yang tinggi dan konsisten, sehingga perlu menjadi prioritas utama dalam pengawasan kualitas udara. Sementara itu, kelompok Medium Risk menunjukkan kondisi yang fluktuatif, dan kelompok Low Risk relatif memiliki kualitas udara yang lebih baik.

### BINNING

Mengelompokkan nilai PM2.5 ke dalam interval kualitas udara agar data mudah dibaca, bisa dibandingkan antar station, musim, dan waktu, serta terlihat seberapa sering tiap tingkat polusi terjadi. proses kategorinya telah dilakukan dan didapatkan kolom pm25_category

```python
# Distribusi Hasil Binning

bin_dist = df['pm25_category'].value_counts().reindex(
    ['Baik','Sedang','Buruk','Tidak Sehat','Parah','Berbahaya']
)

bin_dist
```

```python
# Analisis Proporsi Data Berbahaya per Station

danger = df[df['pm25_category'].isin(['Tidak Sehat','Parah','Berbahaya'])]

danger_prop = (
    danger.groupby('station').size() /
    df.groupby('station').size()
).reset_index(name='proporsi_berbahaya')

danger_prop = danger_prop.sort_values(by='proporsi_berbahaya', ascending=False)
danger_prop
```

```python
# Visualisasi Proporsi Data Berbahaya per Station

plt.figure(figsize=(10,6))

sns.barplot(
    data=danger_prop,
    x='proporsi_berbahaya',
    y='station'
)

plt.title('Proporsi Kondisi Tidak Sehat–Berbahaya per Station')
plt.xlabel('Proporsi (%)')
plt.ylabel('Station')

# ubah ke persen di label
for i, v in enumerate(danger_prop['proporsi_berbahaya']):
    plt.text(v + 0.005, i, f'{v*100:.1f}%', va='center')

plt.tight_layout()
plt.show()
```

**INSIGHT:**

Analisis binning menunjukkan bahwa kualitas udara cenderung berada pada kondisi sedang hingga buruk, dengan kejadian tidak sehat hingga berbahaya yang cukup sering muncul. Terdapat perbedaan antar titik pemantauan, di mana beberapa lokasi lebih sering mengalami kondisi berisiko tinggi. Selain itu, pola waktu juga terlihat jelas, dengan polusi yang lebih tinggi pada musim dingin dan malam hari. Secara keseluruhan, kualitas udara dipengaruhi oleh lokasi, musim, dan waktu, serta menunjukkan bahwa kondisi berbahaya bukan merupakan kejadian yang jarang.
